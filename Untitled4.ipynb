```python
from pyspark.ml.feature import VectorAssembler
```

```python
from pyspark.ml.regression import LinearRegression
```

```python
import pyspark.sql.functions as F
```

```python
spark = SparkSession.builder.appName("LR-pyspark").getOrCreate()
```

```python
data=[
    (3.5, 5.8), 
    (6.7, 9.7), 
    (9.8, 11.2), 
    (14.12, 16.8), 
    (19.0, 25.0)
]
```

```python
df= spark.createDataFrame(data, schema=["x", "y"])
```

```python
df.show()
```

```python
assembler = VectorAssembler(inputCols= ['x'], outputCol="features")
```

```python
df_features= assembler.transform(df).select("features", 'x', 'y')
```

```python
df_features.show()
```

```python
lr = LinearRegression(featuresCol="features", labelCol='y')
model = lr.fit(df_features)
```

```python
intercept = model.intercept
print(intercept)
```

```python
slope=float(model.coefficients[0])
print(slope)
```

```python
r2= model.summary.r2
print(r2)
```

```python
pred= model.transform(df_features).withColumnRenamed("Prediction", "y_hat").withColumn("residual", F.col('y')-F.col('y_hat'))
```

```python
pred.show()
```

```python
pred.select('x', 'y', 'y_hat', 'residual').show()
```

```python
pdf = pred.select('x', 'y', 'y_hat', 'residual').toPandas()
```

```python
print(np)
```

```python
import numpy as np
import matplotlib.pyplot as plt

x_vals = pdf['x'].values
y_vals = pdf['y'].values

x_line = np.linspace(x_vals.min(), x_vals.max(), 200)
y_line = intercept + slope * x_line

plt.figure(figsize=(10,8))
plt.scatter(x_vals, y_vals)
plt.plot(x_line, y_line)
plt.show()
```
